## Basic Idea
- Vocabulary = [1st, 2nd, 3rd...]
- Decoder outputs probabilities for each token
- Highest probability token is selected

```
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

prompt = "once upon a time"
input_ids = tokenizer(prompt, return_tensors="pt")

gen_tokens = model.generate(**input_ids)
print(tokenizer.decode(gen_tokens[0]))
```

## Max New Tokens

```
gen_tokens = model.generate(**input_ids, max_new_tokens=100)
print(tokenizer.decode(gen_tokens[0]))
```

## No Repeat Ngram

```
gen_tokens = model.generate(**input_ids, max_new_tokens=100, no_repeat_ngram_size=2)
print(tokenizer.decode(gen_tokens[0]))
```

## Beam Search

```
Image("/content/beam_search.png", width=300)

gen_tokens = model.generate(**input_ids, num_beams=2)
print(tokenizer.decode(gen_tokens[0]))
```

## Sampling

```
gen_tokens = model.generate(**input_ids, max_new_tokens=100, do_sample=True)
print(tokenizer.decode(gen_tokens[0]))
```

## Top-K Sampling

```
gen_tokens = model.generate(**input_ids, do_sample=True, top_k=5)
print(tokenizer.decode(gen_tokens[0]))
```

## Top-P Sampling

```
gen_tokens = model.generate(**input_ids, do_sample=True, top_p=0.1)
print(tokenizer.decode(gen_tokens[0]))
```

## Temperature

```
gen_tokens = model.generate(**input_ids, do_sample=True, temperature=0.001)
print(tokenizer.decode(gen_tokens[0]))

gen_tokens = model.generate(**input_ids, do_sample=True, temperature=0.9)
print(tokenizer.decode(gen_tokens[0]))
```

## Softmax with Temperature

```
import numpy as np

logits = np.array([2.0, 1.0, 0.1])

def softmax(logits):
    return np.exp(logits) / np.sum(np.exp(logits))

def softmax_with_temp(logits, temp):
    return np.exp(logits / temp) / np.sum(np.exp(logits / temp))

print(softmax(logits))
print(softmax_with_temp(logits, 0.01))
print(softmax_with_temp(logits, 1.8))
```
